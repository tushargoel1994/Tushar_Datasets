# Microsoft Financial Sample.xlsx - Basic Analysis

## Basic Analysis

### Revenue by segment and country heatmap (pivot table, segmentation)
##### Pivot Gross sales by segment X Country. Identify top 3 revenue generating combos and flag underperformers.

### Monthly Revenue trend line (Time series, Seasonality)
##### Plot sales by month across year. Spot seasonality peak and troughs

### Gross Margin % by Product (Margin, Ranking)
##### Compute (Gross Sales - COGS)/ Gross Sales per product. Rank product by margin efficiency

### Discount Impact Waterfall (Waterfall, Discount Analysis)
##### Build a waterfall: Gross Sales -> Subtract discounts -> arrive at Net Sales. Show by Discount bank

### Unit sold vs Profit Scatter (Scatter, Outlier Detection)
##### Plot units sold on X vs Profit on Y, color by segment. Look for volume profit anomalies.

### Top 3/ Bottom 3 Products by Profit (Ranking)
##### Group by product, sum profit, sort ascending and descending. Present as bar chart comparison

### YoY growth by Segment (Growth Rate)
##### Compare sales in Year 1 vs Year 2 per segment. Compute % growth and flag shrinking segments

### COGS as % sales by country (Cost Efficiency, Geographic variability)
##### COGS/Sales ratio by country surfaces operational cost differences across geographies

### Reveneu Mix (% total) Donut Chart (Mix Analysis & Concentration Risk)
##### Show each segment share of total gross sales as donut. Identify Concetration risk.

### Discount Band Profitability table (Discount Strategy & Profitability)
##### Group by discount band -> show average sales, profit, margin %. Are higher discounts destroying value?

### Regression: Price elasticity of demand (OLS Regression, Pricing Strategy)
##### OLS Regression of units sold on sale price per product. Estimate price elasticity. Flag Inelastic Product (Pricing Power)

### ANOVA: Does discount band significantly affect profit? (ANOVA, Hypothesis testing)
##### Run one way ANOVA across discount band groups on profit. Test if discount tiers produce statistically different profit distribution

### Correlation Matrix of financial Drivers (Correlation, Non-linearity)
##### Pearson correlation across Units Sold, Sale Price, Discounts, COGS, Profit Heatmap + VIF to detect multi-collinearity

### Time Series decomposition of monthly sales (STL decomposition, seasonality)
##### Use statsmodels STL decomposition to separate trend, seasonal and residual components from monthly sales

### Monte-carlo simulation (Monte Carlo, Risk Range)
##### Sample Units sold and sale price from fitted distribution

### Contribution Margin and Breakeven Analysis (Contribution Margin, Sensistivity Analysis)
##### Calculate Contribution Margin = Sales - Manufacturing Price. Compute Breakeven units per product. Build sensisitivity table by changing price

### Segment P&L Waterfall - Executive View (P&L, CFO Desk)
##### For each segment: Gross Sales -> Discounts -> Net Sales -> COGS -> Gross Profit -> Profit. Present a side-by-side waterfall. Teh standard CFO deck format
